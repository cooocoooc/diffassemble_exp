## 2. File operation:
- create directory
  - mkdir
- open folder(win)/directory(linux)
  - cd folder_path

- view all files on the current folder/directory
  - win: dir 
  - linux & mac: ls
- move file
 - linux & mac
   - input: `mv file target_path`
     - e.g. relative path
       - file
       - path
       - - ->child_path
       - - child path: `mv ../../file .`
       - `mv file path/child_path`
   - move multiple files:
     - input: `mv file1 file2 target_path`
   - move all of the files:
     - input: `mv ./* target_path`
       - using `*`
 - win: 
   - input: `move file target_path`
     - mv -> move
     - / -> \
     - e.g. relative path
       - file
       - path
       - - ->child_path
       - - child path:` move ..\..\file .`
       - `move file path\child_path`
   - move multiple files:
     - input: ~~`move file1 file2 target_path`~~
       - error: The syntax of the command is incorrect
       - fix: only one file at a time
   - move all of the files:
     - input: `move .\* target_path`
       - using `*`
