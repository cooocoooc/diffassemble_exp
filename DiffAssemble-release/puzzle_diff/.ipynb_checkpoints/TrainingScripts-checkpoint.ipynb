```python
import wandb

from pytorch_lightning.loggers import WandbLogger

wandb.login()  # if logined else wandb.login(key="your_api_key")

run = wandb.init(project="dissassemble-test", entity="cocotea-durham-university")

os.environ["WANDB_RUN_ID"] = run.id

wandb_id = os.environ.get("WANDB_RUN_ID")

wandb_finish_not_call = True

if wandb_finish_not_call:
    '''1. what will happen
          - wandb.finish() automatically calls at the ends of training
            - Pytorch Lightning
          - others:
            i. stay in "running" state W&B dashboard
            ii. logs or artifacts not be uploaded properly
                e.g. data in W&B buffers
            iii. resources not be released
                 e.g. threads and processes hanging,
                     especially in long-running scripts or notebooks
       2. what to do next
          i. check on W&B dashboard
             go to "https://wandb.ai" and check the run state of the project:
             -finished
             -crashed
             -running(temporarily)
          ii. resume the run    
    '''

    # if wandb_id is saved, e.g.
    # wandb_id = "b34q8gp6"
    if wandb_id:
        '''
        how the parametre "resume" works:
        i. "allow"            
           - is a safely resume training without considering the states of the run 
           - a run with wandb_id exists -> continues to the same run
           - no such run -> create a new run with this ID
           - no ID provied -> no run
        ii. "must"
            - resume only if the run exists
        iii. "never"
            - always create a new run,even if a run with the ID exists
        '''
        wandb_logger = WandbLogger(project="dissassemble-test",id = wandb_id,resume = "allow")
        
        # Access the W&B run ID
        run_id = wandb_logger.experiment.id
        print(f"W&B Run ID: {run_id}")
   
else:
    wandb.finish()
```
